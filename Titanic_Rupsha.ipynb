```python
titan['family']=titan['sibsp']+titan['parch']
titan['embarked']=titan['embarked'].fillna('S')
titan['age']=titan['age'].fillna(titan['age'].mean)
titan.drop(['sibsp', 'parch', 'adult_male', 'embark_town', 'deck', 'who', 'alone', 'alive'], axis='columns', inplace=True)
titan['sex'].replace(['male', 'female'], [0,1], inplace=True)
titan['embarked'].replace(['S','C','Q'], [0,1,2], inplace=True)
titan['class'].replace(['First', 'Second', 'Third'], [1,2,3], inplace=True)
titan.info()
```

```python
titan.corr()
```

```python
titan.cov()
```

```python
from sklearn.model_selection import train_test_split

x=titan[['pclass', 'family', 'sex', 'class', 'embarked']]
y=titan[['survived']]
  
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier()
classifier.fit(x_train, y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
